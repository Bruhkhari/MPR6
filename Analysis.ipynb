```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('athlete_events.csv')
region_df = pd.read_csv('noc_regions.csv')
```

```python
#Replacing NOC three letter codes with respective regions
```

```python
df = df.merge(region_df,on='NOC',how='left')
```

```python
df.drop_duplicates(inplace=True)
```

```python
#removing NaN values from medals by one hot encoding
```

```python
df['Medal'].value_counts()
```

```python
pd.get_dummies(df['Medal'])
```

```python
#combining the dataframes
```

```python
df = pd.concat([df,pd.get_dummies(df['Medal'])],axis=1)
```

```python
df.groupby('NOC').sum()[['Gold','Silver','Bronze']].sort_values('Gold', ascending=False).reset_index()
```

```python
#As the dataset is medals by athletes, the team sports have bloated value for countries.
#Hence we will drop those duplicated team medals for our Medal tally section.
```

```python
medal_tally = df.drop_duplicates(subset =['Team','NOC','Games','Year','City','Sport','Event','Medal'])
```

```python
medal_tally.groupby('NOC').sum()[['Gold','Silver','Bronze']].sort_values('Gold', ascending=False).reset_index()
```

```python
medal_tally = medal_tally.groupby('NOC').sum()[['Gold','Silver','Bronze']].sort_values('Gold', ascending=False).reset_index()
```
